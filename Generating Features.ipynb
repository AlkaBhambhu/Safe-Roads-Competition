```python
import pandas as pd
import numpy as np
import seaborn as sns
import util
from scipy import stats
```

```python
pd.set_option('max_columns', None)
```

```python
ksi_df = pd.read_csv('Data/KSI.csv')
```

```python
# I AM ONLY LOOKING AT AGE AND BEHAVIOUR FOR DRIVERS
# NEGLECTING PASSENGERS, PEDESTRIANS, ETC.

ksi_df = ksi_df[ksi_df['INVTYPE'].str.contains('Driver')]
```

```python
ksi_df['LATITUDE'] = ksi_df['LATITUDE'].apply(util.truncate, args = (3, ))
ksi_df['LONGITUDE'] = ksi_df['LONGITUDE'].apply(util.truncate, args = (3, ))
```

```python
# dont run this, it just makes life harder

#ksi_df.replace('<Null>', np.nan, inplace = True)
```

```python
# interaction variables between boolean columns and actor columns
# , 'PEDACT', 'PEDCOND' , 'CYCACT', 'CYCCOND'
```

```python
# its ugly, its stupid, buts its mine, and it works

l = []
for col in ['INVAGE', 'WARDNUM', 'RDSFCOND', 'VISIBILITY', 'LIGHT', 'TRAFFCTL', 'VEHTYPE', 'DRIVACT', 'DRIVCOND', 'SPEEDING', 'AG_DRIV', 'ALCOHOL', 'DISABILITY', 'REDLIGHT', 'MANOEUVER']:    
    l.append(ksi_df.groupby(['LATITUDE', 'LONGITUDE'])[col].apply(lambda x: x.mode().iloc[0]).reset_index())
```

```python
# who cares about vectorization?
# loop like a neanderthal

looper = l[0]
for df in range(1, len(l)):
    looper = looper.merge(l[df], on = ['LATITUDE', 'LONGITUDE'], how = 'inner')
```
