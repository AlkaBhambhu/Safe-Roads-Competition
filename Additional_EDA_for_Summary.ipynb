```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ksi = pd.read_csv('KSI.csv')
```

```python
ksi[ksi['INVAGE'] != 'unknown'].groupby('INVAGE')['ACCNUM'].nunique().sort_values().plot(kind = 'barh', figsize = (8,5))
```

```python
plt.figure(figsize = (5,10))
sns.countplot(data = ksi[(ksi['MANOEUVER'] != '<Null>') & (ksi['DRIVACT'] != '<Null>') & (ksi['DRIVACT'] != 'Driving Properly')], y = 'MANOEUVER', hue = 'DRIVACT')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
plt.figure(figsize = (10,15))
sns.countplot(data = ksi[(ksi['IMPACTYPE'] != '<Null>')], y = 'HOUR', hue = 'IMPACTYPE')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
plt.figure(figsize = (5,10))
sns.countplot(data = ksi[(ksi['IMPACTYPE'] != '<Null>') & (ksi['DRIVCOND'] != '<Null>') & (ksi['DRIVCOND'] != 'Unknown')], y = 'IMPACTYPE', hue = 'DRIVCOND')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
plt.figure(figsize = (5,10))
sns.countplot(data = ksi[(ksi['DRIVCOND'] != '<Null>') & (ksi['DRIVCOND'] != 'Unknown') & (ksi['DRIVACT'] != 'Null')], y = 'DRIVACT', hue = 'DRIVCOND')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
ksi[ksi['INJURY'] == 'Fatal'].groupby('INVAGE')['ACCNUM'].nunique().plot(kind='bar')
```

```python
injury_by_age = ksi.groupby(['INVAGE','ACCLASS'])['ACCNUM'].nunique()
injury_by_age.groupby(level=[0]).apply(lambda g: 100* g / g.sum()).round(2)
```

```python
ksi[ksi['VISIBILITY'] != '<Null>'].groupby('VISIBILITY')['ACCNUM'].nunique().plot(kind='bar')
```

```python
ksi[ksi['RDSFCOND'] != '<Null>'].groupby('RDSFCOND')['ACCNUM'].nunique().plot(kind='bar')
```
