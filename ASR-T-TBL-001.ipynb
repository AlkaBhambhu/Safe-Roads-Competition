```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_sas('traffic_collisions_asr_t_tbl_001.sas7bdat', format = 'sas7bdat', encoding = 'iso-8859-1')
```

```python
months = ["January", "February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]
df['Month'] = pd.Categorical(df['Month'], categories = months, ordered = True)
```

```python
df.sort_values(['Year', 'Month'], ascending = True).groupby(['Year', 'Month'])['EventUniqueId'].nunique().unstack().plot(kind = 'bar', stacked = True, figsize = (20, 5))
```

```python
an = df.sort_values(['Month', 'Year'], ascending = True).groupby(['Year', 'Month'])['EventUniqueId'].nunique()
p1 = an.plot(kind = 'line', figsize = (20, 5))
p1.set_xticks(range(len(an)));
p1.set_xticklabels(["%02d-%s" % item for item in an.index.tolist()], rotation=90);
```

```python
df.sort_values(['Hour', 'Year'], ascending = False).groupby(['Hour', 'Year'])['EventUniqueId'].nunique().unstack().plot(kind = 'line', xticks = range(24))
```

```python
def seasons(col):
    if col['Month'] in ('March', 'April', 'May'):
        return 'Spring'
    if col['Month'] in ('June', 'July', 'August'):
        return 'Summer'
    if col['Month'] in ('September', 'October', 'November'):
        return 'Fall'
    if col['Month'] in ('December', 'January', 'February'):
        return 'Winter'
```

```python
df['Season'] = df.apply(seasons, axis = 1)
```

```python
df['Season'].value_counts().sort_values(ascending = False).plot(kind = 'barh')
```
